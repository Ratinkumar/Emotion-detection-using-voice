```python
import os
import shutil
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
```

```python
Emotions = ['anger', 'disgust', 'fear', 'happiness', 'neutral', 'sadness','surprise']

for e in Emotions:
    parent_dir = "/content/drive/MyDrive/Project/VisualData/"
    path = os.path.join(parent_dir, e)

    # Check if the directory already exists
    if not os.path.exists(path):
        os.mkdir(path)
```

```python
#n_fft is the number of points in each short-time Fourier transform (STFT) window.
#It determines the size of the window function used to analyze the signal.
n_fft=2048
#hop_length is the number of samples between successive frames in the STFT.
#It determines the amount of overlap between consecutive windows.
hop_length=512
#n_mels is the number of Mel filterbanks to use in the Mel spectrogram computation.
#It defines the number of output channels or frequency bins in the Mel spectrogram.
n_mels=128
```

```python
#storing the MelSpectrogram of all the emotions

dirs = os.listdir("/content/drive/MyDrive/Project/Dataset")

for d in dirs :

    src = "/content/drive/MyDrive/Project/Dataset/" + d + '/'
    dest = "/content/drive/MyDrive/Project/VisualData/" + d + '/'
    files = os.listdir("/content/drive/MyDrive/Project/Dataset/" + d)

    for f in files :
        y, sr = librosa.load(src + f)
        emotion, _ = librosa.effects.trim(y)
        S = librosa.feature.melspectrogram(y=emotion, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels)
        S_DB = librosa.power_to_db(S, ref=np.max)
        librosa.display.specshow(S_DB, sr=sr, hop_length=hop_length, x_axis='time', y_axis='mel');
        plt.colorbar(format='%+2.0f dB');
        plt.savefig(dest + f[:-4] + '.png')
        plt.clf()
```
