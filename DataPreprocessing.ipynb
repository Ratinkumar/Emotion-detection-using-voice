```python
#These library will help in moving files from one directory to another
import os
import shutil
```

```python
Emotions = ['anger', 'disgust', 'fear', 'happiness', 'neutral', 'sadness', 'surprise']

#Here, i am making directory for each emotion
for e in Emotions:
  path="/content/drive/MyDrive/Project/Dataset/" + e
  os.mkdir(path)
```

```python
directory = os.listdir("/content/drive/MyDrive/Project/AudioData")
os.mkdir("/content/drive/MyDrive/Project/Dataset/Temporary")
tempdir = "/content/drive/MyDrive/Project/Dataset/Temporary/"

counters = [0, 0, 0, 0, 0, 0, 0]
#Here, i am transferring data from original dataset to different emotion directory
for d in directory:
  src = "/content/drive/MyDrive/Project/AudioData/" + d + '/'
  dest = "/content/drive/MyDrive/Project/Dataset/"
  files = os.listdir("/content/drive/MyDrive/Project/AudioData/" + d)

  for f in files :
        if f[0] == 'a' :
          counters[0] = counters[0]+1
          shutil.copy(src + f, tempdir)
          os.rename(tempdir + f, tempdir + f[0] + str(counters[0]) + '.wav')
          shutil.move(tempdir + os.listdir(tempdir)[0] , dest + "anger/")

        elif f[0] == 'd' :
          counters[1] = counters[1]+1
          shutil.copy(src + f, tempdir)
          os.rename(tempdir + f, tempdir + f[0] + str(counters[1]) + '.wav')
          shutil.move(tempdir + os.listdir(tempdir)[0] , dest + "disgust/")

        elif f[0] == 'f' :
          counters[2] = counters[2]+1
          shutil.copy(src + f, tempdir)
          os.rename(tempdir + f, tempdir + f[0] + str(counters[2]) + '.wav')
          shutil.move(tempdir + os.listdir(tempdir)[0] , dest + "fear/")

        elif f[0] == 'h' :
          counters[3] = counters[3]+1
          shutil.copy(src + f, tempdir)
          os.rename(tempdir + f, tempdir + f[0] + str(counters[3]) + '.wav')
          shutil.move(tempdir + os.listdir(tempdir)[0] , dest + "happiness/")

        elif f[0] == 'n' :
          counters[4] = counters[4]+1
          shutil.copy(src + f, tempdir)
          os.rename(tempdir + f, tempdir + f[0] + str(counters[4]) + '.wav')
          shutil.move(tempdir + os.listdir(tempdir)[0] , dest + "neutral/")

        elif f[:2] == 'sa' :
          counters[5] = counters[5]+1
          shutil.copy(src + f, tempdir)
          os.rename(tempdir + f, tempdir + f[:2] + str(counters[5]) + '.wav')
          shutil.move(tempdir + os.listdir(tempdir)[0] , dest + "sadness/")

        elif f[:2] == 'su' :
          counters[6] = counters[6]+1
          shutil.copy(src + f, tempdir)
          os.rename(tempdir + f, tempdir + f[:2] + str(counters[6]) + '.wav')
          shutil.move(tempdir + os.listdir(tempdir)[0] , dest + "surprise/")

os.rmdir(tempdir)
```
